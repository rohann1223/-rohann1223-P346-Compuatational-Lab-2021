```python
#converts a simple list of length N to 1 x N matrix
#input is a list X

def mat_converter(X):
    if isinstance(X[0], float): #gives back the data type of the variable
        x = []
        x.append([])
        for i in X:
            x[0].append(i)
        return x
    else:
        return x
```

```python
#To do operation of matrix transpose

def mat_transpose(X):
    Z = []
    X = mat_converter(X)
    for i in range(0,len(X[0])):
        z = []
        Z.append(z)
        for j in range(0,len(X)):
            Z[i].append(X[j][i])
    return Z

#To do the Matrx Multipilaction

def mat_multiplication(X,Y):
    Z = []
    X = mat_converter(X)
    Y = mat_converter(Y)
    
    if len(X[0]) == len(Y):
        
        for i in range(0,len(X)):
            empty_list = []
            Z.append(empty_list)
            
            for j in  range(0,len(Y[0])):
                Z[i].append(0)
                Z[i][j] = 0
                for k in range(0,len(X[0])):
                    Z[i][j] = Z[i][j] + (X[i][j] * Y[k][j])
                    
        if len(Z) == 1 and len(Z[0]) == 1:
            Z = Z[0][0]
            return Z
        
        else:
            return Z
    else:
        print('the given matrices cannot be multiplied because of dimension problem')
            
            
```
