### Q1-Defining matrices, storing them, finding C.D, AB, CA, and BD

```python
#Defining Matrices A,B,C,D in their respective csv files and manipulating them.

import pandas as pd   #to read csv files ONLY
import numpy as np    #to store matrices ONLY
#--------------------------------
A=np.array(pd.read_csv("A.csv",header=None))
B=np.array(pd.read_csv("B.csv",header=None))
C=np.array(pd.read_csv("C.csv",header=None))
D=np.array(pd.read_csv("D.csv",header=None))
#--------------------------------
nrow = lambda x: x.shape[0]
ncol = lambda x: x.shape[1]
#--------------------------------
print(A)
def product(X,Y):
    if ncol(X)!=nrow(Y):     #to check for incompatible matrices
        raise ValueError("Error! ncol(X) doesn't match nrow(Y). Check for product compatibility.")
    else:
        XY=np.array(np.zeros((nrow(X),ncol(Y)), dtype = np.float))
        for i in range(0,nrow(X)):
            for j in range(0,nrow(Y)):
                for k in range(0,ncol(Y)):
                    XY[i,k]=XY[i,k]+X[i,j]*Y[j,k]
        return XY
def transpose(X):        #to calculate the transpose of X
    Xt=np.array(np.zeros((ncol(X),nrow(X)), dtype = np.float))
    for i in range(0,nrow(X)):
        for j in range(0,ncol(X)):
            Xt[j,i]=X[i,j]
    return Xt

print("Dot Product of C and D is:\n",product(C,transpose(D)))
print("")
print("Matrix Product of A and B is:\n",product(A,B))
print("")
print("Matrix Product of C and A is:\n",product(C,A))
print("")
print("Matrix Product of B and D is:\n",product(B,transpose(D)))
```

```python
from math import atan #to define phase angle

class myComplex():

    def __init__(self, real, img, real2, img2): #class has to be defined with 4 numbers to make two complexs have real and img part
        self.r = real
        self.i = img
        self.r2 = real2
        self.i2 = img2
        
    def show(self): #to show complex numbers which are given
        print ("The complex numbers entered by you are " + str(self.r)+ " + " + str(self.i)+ "i" + " and " + str(self.r2)+ " + " + str(self.i2)+ "i")
        
        
    def add(self): #to add the two numbers
        print ("The sum is " + str(self.r + self.r2) + " + " + str(self.i +self.i2) +"i")  
        
    def product(self): #to multiply the two numbers
        print ("The product is " + str((self.r * self.r2) - (self.i * self.i2)) + " + " + str((self.i * self.r2) + (self.r * self.i2)) +"i")
        
        
    def division(self): #to divide the two numbers
        if self.r ==0 and self.i ==0 and self.r2==0 and self.i2 ==0:
            print('The division is not defined')
        else:
            print("The division of the two complex numbers are " + str(((self.r * self.r2) - (self.i * self.i2))/((self.r2)**2 - (self.i2)**2)) + " + " + str(((self.i * self.r2) + (self.r * self.i2))/((self.r2)**2 - (self.i2)**2)) +"i")

    def conjugate(self): #to get conjugate of both numbers
        print( "The conjugate of " + str(self.r)+ " + " + str(self.i)+ "i = " + str(self.r)+ " - " + str(self.i)+ "i")
        print( "The conjugate of " + str(self.r2)+ " + " + str(self.i2)+ "i = " + str(self.r2)+ " - " + str(self.i2)+ "i")

    def modulus(self): #to get mod
        print("The modulus of " + str(self.r)+ " + " + str(self.i)+ "i = " + str(((self.r)**2+(self.i)**2)**0.5))
        print("The modulus of " + str(self.r2)+ " + " + str(self.i2)+ "i = " + str(((self.r2)**2+(self.i2)**2)**0.5))

    def phase(self): #for defining phase
        if self.r ==0 and self.i ==0 and self.r2==0 and self.i2 ==0:
            print('The tan of the complex number is not defined')
        else:
            print("The phase angle of " + str(self.r)+ " + " + str(self.i)+ "i = " + str(atan(self.r/self.i)))
            print("The phase angle of " + str(self.r2)+ " + " + str(self.i2)+ "i = " + str(atan(self.r2/self.i2)))
   
 #outputs
complex = myComplex(1,2,3,4)
complex.show()
complex.add()
complex.product()
complex.division()
complex.conjugate()
complex.modulus()
complex.phase()
print()
complex = myComplex(12,14,6,8)
complex.show()
complex.add()
complex.product()
complex.division()
complex.conjugate()
complex.modulus()
complex.phase()    
print()
complex = myComplex(0,1,0,1)
complex.show()
complex.add()
complex.product()
complex.division()
complex.conjugate()
complex.modulus()
complex.phase() 
print()
complex = myComplex(0,0,0,0)
complex.show()
complex.add()
complex.product()
complex.division()
complex.conjugate()
complex.modulus()
complex.phase() 
print()
complex = myComplex(-1,2,2.5,3)
complex.show()
complex.add()
complex.product()
complex.division()
complex.conjugate()
complex.modulus()
complex.phase()        
        
        
```

### Q3- calculates average distance between any 2 points in a straight line of N equidistant points

```python
def avg_dist(num_points):

    list_of_distances=[]
    sum=0 
    if num_points < 1 or num_points > 0 and num_points - int(num_points)!= 0 :
        print("The entered value is not a natural number.")
    else:
        for i in range(1,num_points+1): #chooses a point P
            for n in range(1,num_points+1): #chooses another point Q to measure distance from P
                distance= abs(i-n)
                list_of_distances.append(distance) #makes a list of distances

        for element in list_of_distances: 
            sum+=element                      #sums all the distances
        average_distance=sum/len(list_of_distances)  #divides sum by total number of point pairs considered 
        print("Average distance between two points on a straight line with",num_points,"equidistant points is ",average_distance,'units.')

avg_dist(1)
avg_dist(0)
avg_dist(-3)
avg_dist(3.2)
avg_dist(3)
avg_dist(5)
```
